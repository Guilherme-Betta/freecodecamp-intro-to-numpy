### Create a 3*3 identity numpy matrix with ones on the diagonal and zeros elsewhere.

```python
# your code goes here
id_matrix = np.zeros([3,3])

for x in range(len((id_matrix))):
    for y in range(len((id_matrix))):
        if x == y:
            id_matrix[x, y] = 1

id_matrix   # Não conhecia o método "identity", mas gostei da minha solução.
```

### Order (sort) the given numpy array

```python
# your code goes here
X = [0, 10, -5, 40, -3]

Y = np.copy(X)

for i in range(len(Y+1)):
    # if i == 0:
    #     Y_max = X[i]
    # else:
    #     if Y[i] >= X[i]:
    #         Y[i] = Y_max
    if Y[i+1] <= X[i]:
        Y[i] = Y[i+1]

Y
```
